```python
import seaborn as sns
import pandas as pd

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Print column names to verify
print(titanic.columns)

# Check the first few rows of the DataFrame to ensure it is loaded correctly
print(titanic.head())

# Create the box plot
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
```

```python
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)
```

```python
import warnings
import seaborn as sns
import pandas as pd

# Load the flights dataset from seaborn
flights = sns.load_dataset("flights")

# Suppress warnings in this block
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=FutureWarning)
    # Group by 'year' and sum the 'passengers' column, avoiding the warning
    new = flights.groupby('year').sum().reset_index()

# Create the line plot
sns.lineplot(x='year', y='passengers', data=new)
```
